```python
import os, random
def writeFile(fn, allfiles, v):
    f = open(fn, 'w', encoding='GBK' )
    for imagelabel in allfiles: 
            f.write(f'{imagelabel[0]} {v[imagelabel[1]]}\n')
    f.close()                         
def getfiles(path, tag):
        allfiles = []
        labels = []
        for root, dirs, files in os.walk(path):            
            for name in files:
                if name[-3:].lower() not in ['jpg','png',]: continue
                parts = os.path.normpath(root).split(os.sep)
                #print(root, dirs,name, parts)
                allfiles.append([os.path.join(root, name),parts[1]])              
                labels.append(parts[1])
        # 去重并保持原始顺序
        unique_items = list(dict.fromkeys(labels))
        # 创建元素到编号的字典
        values = {item: idx for idx, item in enumerate(unique_items)}
        print(values)
        f = open(tag+'labels.txt', 'w', encoding='GBK' )
        for k, v in values.items(): 
            f.write(f'{k} {v}\n')
        f.close()                         
        # 打乱列表顺序
        random.shuffle(allfiles)
        n = len(allfiles)
        a1, a2 = int(n*.9), int(n*1)
        writeFile(tag+'train.txt', allfiles[0:a1], values)
        writeFile(tag+'val.txt', allfiles[a1:a2], values)
        writeFile(tag+'test.txt', allfiles[a2:n], values)

getfiles('dataset', 'sat')
```
